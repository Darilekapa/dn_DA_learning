1) Создайте функцию student_report(grades),
которая принимает словарь формата:

grades = {"Анна": 85, "Иван": 50, "Мария": 91, "Павел": 62}


и возвращает словарь:

{
  "average": ...,         # средняя оценка
  "best_student": ...,    # имя с наивысшей оценкой
  "passed": ...,          # сколько студентов набрали >= 70
  "failed": ...           # сколько студентов набрали < 70
}

```python
def student_report(grades):
   
    avg= sum(grades.values())/len(grades.values())
    max_temp=0
    best_day=''
    s_70_b=[]
    s_70_m=[]   
    for name,kol in grades.items():
        if kol > max_temp:
            max_temp = kol
            best_day = name
       
        if kol>=70:
            s_70_b.append(name)
        if kol<70:
            s_70_m.append(name)
                   
    result={
  "average": avg,         # средняя оценка
  "best_student": best_day,    # имя с наивысшей оценкой
  "passed": len(s_70_b),          # сколько студентов набрали >= 70
  "failed": len(s_70_m)           # сколько студентов набрали < 70
}
    return result
grades = {"Анна": 85,
          "Иван": 50, 
          "Мария": 91,
          "Марина": 93,
          "Павел": 62
          }

print (student_report(grades))
```

2. Создайте колонку Month (возьмите месяца из Order Date) и постройте line chart,
где ось X — месяц, ось Y — сумма продаж за этот месяц. (понадобиться группировка по месяцам)

```python
df_sss['Order Date Month'] = df_sss["Order Date"].dt.month_name()
```

```python
df_sss
```

```python
# 3. Группируем по месяцам и считаем сумму продаж
monthly_sales = df_sss.groupby('Order Date Month')['Sales'].sum()
#monthly_s = monthly_sales.nlargest(12,'Order Date Month')
print("\nСумма продаж по месяцам:")
print(monthly_sales)
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']

monthly_sales = monthly_sales.reindex(month_order)
print("\nСумма продаж по месяцам (в хронологическом порядке):")
print(monthly_sales)
```

```python

# 4. Строим линейный график
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index,
         monthly_sales.values,
         marker='o', 
         linewidth=2,
         markersize=8, 
         markerfacecolor='red')

# Настройки графика
plt.title('Динамика продаж по месяцам', fontsize=14, fontweight='bold')
plt.xlabel('Месяц', fontsize=12)
plt.ylabel('Сумма продаж ($)', fontsize=12)
plt.grid(True, alpha=0.3)


# Добавляем подписи значений на точках
for i, (month, sales) in enumerate(zip(monthly_sales.index,
         monthly_sales.values,)):
    plt.annotate(f'${sales:,.0f}', 
                (month, sales),
                textcoords="offset points",
                xytext=(0,10),
                ha='center',
                fontsize=9)

plt.tight_layout()
plt.show()
```
